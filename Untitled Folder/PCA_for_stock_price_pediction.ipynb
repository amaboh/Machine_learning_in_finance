```python
# import neccasary librareries
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
import bs4 as bs
```

```python
#Setting the plot style
plt.style.use('fivethirtyeight')
```

```python
# Function to scrape the tickers of companies in the S&P 500 form wikipedia using beautiful soup

def save_sp500_tickers():

    resp = requests.get('http://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
    soup = bs.BeautifulSoup(resp.text, 'html')
    table = soup.find('table', {'class': 'wikitable sortable'})
    tickers = []
    for row in table.findAll('tr')[1:]:
        ticker = row.findAll('td')[0].text
        if not '.' in ticker:
            tickers.append(ticker.replace('\n',''))
        
    return tickers

tickers = save_sp500_tickers()
    
```

```python
#Using the Yahoo finance library we download the adjusted close price of the S&P500 stocks price

prices = yf.download(tickers, start='2019-01-01')['Adj Close']
```

```python
#Calculate the daily returns of the S&P500 stocks using logarithm returns method
rs = prices.apply(np.log).diff(1)
```

```python
#Plot the daily returns of the S&P500 stocks 

rs.plot(legend=0, figsize=(10,6), grid=True, title='Daily Returns of companies stock price in the S&P500')
plt.tight_layout()
plt.savefig('tmp.png')
```

```python
(rs.cumsum().apply(np.exp)).plot(legend=0, figsize=(10,6), grid=True, title='Cumulative Returns of the Companies Stocks prices in the S&P500')
plt.tight_layout()
plt.savefig('tmp.png')
```
